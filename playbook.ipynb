# TIME SERIES

```python
#TIME SERIES CONFIRMED

def time_series_confirmed(col): 
    series_confirmed = confirmed.loc[:, confirmed.columns[-col:]]
    region_name_confirmed= confirmed.loc[:,confirmed.columns[1]]
    pd.DataFrame(region_name_confirmed)
    df = pd.concat([series_confirmed, region_name_confirmed], axis=1, join="inner")
    time_series_per_country_confirmed = df.groupby(['Country/Region']).sum()  # same countries ki series values are summed
    
    return time_series_per_country_confirmed

```

```python
#TIME SERIES DEATHS

def time_series_deaths(col):

    series_deaths = deaths.loc[:, deaths.columns[-col :]]
    region_name_deaths= deaths.loc[:,deaths.columns[1]]
    pd.DataFrame(region_name_deaths)
    df = pd.concat([series_deaths, region_name_deaths], axis=1, join="inner")
    time_series_per_country_deaths = df.groupby(['Country/Region']).sum()
    
    return time_series_per_country_deaths

    
```

```python
#TIME SERIES RECOVERED

def time_series_recovered(col):
    
    series_recovered = recovered.loc[:, recovered .columns[-col:]]
    region_name_recovered = recovered.loc[:,recovered.columns[1]]
    pd.DataFrame(region_name_recovered )
    df = pd.concat([series_recovered , region_name_recovered], axis=1, join="inner")
    time_series_per_country_recovered  = df.groupby(['Country/Region']).sum()
    
    return time_series_per_country_recovered

```

```python
def get_stationarity(timeseries):
    
    # rolling statistics
    rolling_mean = timeseries.rolling(window=12).mean()
    rolling_std = timeseries.rolling(window=12).std()
    
    # rolling statistics plot
    original = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(rolling_mean, color='red', label='Rolling Mean')
    std = plt.plot(rolling_std, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    # Dickeyâ€“Fuller test:
    result = adfuller(timeseries[36])
    print('ADF Statistic: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {}'.format(key, value))
```

```python
rolling_mean = df_log.rolling(window=12).mean()
df_log_minus_mean = df_log - rolling_mean
df_log_minus_mean.dropna(inplace=True)
get_stationarity(df_log_minus_mean)
```

```python
rolling_mean_exp_decay = df_log.ewm(halflife=12, min_periods=0, adjust=True).mean()
df_log_exp_decay = df_log - rolling_mean_exp_decay
df_log_exp_decay.dropna(inplace=True)
get_stationarity(df_log_exp_decay)
```

```python
df_log_shift = df_log - df_log.shift()
df_log_shift.dropna(inplace=True)
get_stationarity(df_log_shift)
```
